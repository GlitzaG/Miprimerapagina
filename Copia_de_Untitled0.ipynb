```python
import numpy as np
u= lambda t: np.piecewise(t,t>=0, [1,0])
g= -10 #rango de graficacion
h= 10 #rango de graficación
dt= 0.1
t= np.arange(g,h,dt) #se definen las constantes
u0= u(t-0)
import matplotlib.pyplot as plt #importamos la libreria matplot
plt.xlabel('EJE X') #se usa este comando para mostar el eje x en la grafica
plt.ylabel('EJE Y. Escalón u (t)') #se usa este comando para poner una etiqueta en
plt.figure(1)
plt.title("U(t)") #se usa este comando para poner un titulo en la grafica
plt.plot(t, u0, "g") #aqui esta la funcion de la grafica y la g es para cambiar el color de la grafica
plt.margins(dt) #margenes
plt.grid() #cuadriculas en la grafica
plt.show() #nos muestra la grafica
```

```python
import numpy as np
import matplotlib.pyplot as plt
t0 = np.arange(-2*np.pi,2*np.pi,0.1)
m = np.sinc(t0)  #(sin(pi*x)/pi*x)
plt.plot(t0,m,"m")
plt.xlabel('EJE X')
plt.ylabel('EJE Y')
plt.title('Funciones de ingenieria, funcion Sinc(x)')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
t1 = np.arange(-2*np.pi,2*np.pi,0.01)
m1 = np.sin(t1)/t1 # (sin(x)/x)
plt.plot(t1,m1,"r")
plt.xlabel('EJE X')
plt.ylabel('EJE Y')
plt.title('Funciones de ingenieria, funcion Sa(x)')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def rectangular(x, a):
    return np.where(np.abs(x) < a/2, 1, 0)

# Definir el rango de valores x
x = np.linspace(-10, 10, 1000)

# Definir el ancho de la función rectangular
ancho = 4

# Calcular los valores de la función rectangular para cada valor de x
y = rectangular(x, ancho)

# Graficar la función rectangular
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Función Rectangular')
plt.grid(True)
plt.ylim(-0.5, 1.5)  # Establecer límites en el eje y
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)  # Generar 1000 puntos en el rango de -5 a 5
y = np.sign(x)  # Aplicar la función signo a los puntos generados

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sign(x)')
plt.title('Gráfica de la función signo')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rampa(x):
    return np.maximum(0, x)

x = np.linspace(-10, 10, 100)  # Crear un arreglo de valores de x

# Calcular los valores correspondientes de y
y = rampa(x)

# Graficar la función rampa
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('rampa(x)')
plt.title('Gráfica de la función rampa')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def triangular(x, a, b, c):
    return np.maximum(0, np.minimum((x - a) / (b - a), (c - x) / (c - b)))

# Parámetros de la función triangular
a = 0.2
b = 0.5
c = 0.8

# Rango de valores x
x = np.linspace(0, 1, 100)

# Calcula los valores y de la función triangular para cada valor de x
y = triangular(x, a, b, c)

# Grafica la función triangular
plt.plot(x, y)
plt.title('Función Triangular')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
